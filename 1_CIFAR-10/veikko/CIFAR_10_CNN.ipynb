```python
import os
os.environ["KERAS_BACKEND"] = "tensorflow"

import keras
from keras import layers
from keras.datasets import cifar10
from keras.regularizers import l2

import numpy as np
import matplotlib.pyplot as plt
```

```python
input_shape = (32, 32, 3)

(x_train, y_train), (x_test, y_test) = cifar10.load_data()

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
inputs = keras.Input(shape=input_shape)

x = layers.RandomRotation(0.1)(inputs)
x = layers.RandomZoom(0.1)(x)

x = keras.layers.Conv2D(128, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.0001), padding='same')(x)
x = keras.layers.Conv2D(128, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.0001), padding='same')(x)
x = keras.layers.MaxPooling2D(pool_size=(2, 2))(x)
x = layers.Dropout(0.2)(x)

x = keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.0001), padding='same')(x)
x = keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.0001), padding='same')(x)
x = keras.layers.MaxPooling2D(pool_size=(2, 2))(x)
x = layers.Dropout(0.2)(x)

x = keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.0001), padding='same')(x)
x = keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.0001), padding='same')(x)
x = keras.layers.MaxPooling2D(pool_size=(2, 2))(x)
x = layers.Dropout(0.2)(x)

x = layers.Flatten()(x)

outputs = layers.Dense(10, activation="softmax")(x)
model = keras.Model(inputs=inputs, outputs=outputs, name="cifar10_model")
```

```python
batch_size = 64
epochs = 30

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.2)
```

```python
plt.plot(history.history["loss"], label="Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.grid()
plt.show()
```

```python
test_scores = model.evaluate(x_test, y_test, verbose=2)
print("Test loss:", test_scores[0])
print("Test accuracy:", test_scores[1])

train_scores = model.evaluate(x_train, y_train, verbose=2)
print("Train loss:", train_scores[0])
print("Train accuracy:", train_scores[1])
```
